```python
import matplotlib.pyplot as plt

def dda_line(x1, y1, x2, y2):
    xes, yes = [x1], [y1]
    
    xd = x2 - x1
    yd = y2 - y1
    step = max(abs(xd), abs(yd))
    
    x_inc = xd / step
    y_inc = yd / step
    
    x, y = x1, y1

    for _ in range(step):
        x += x_inc
        y += y_inc
        xes.append(round(x))
        yes.append(round(y))

    return xes, yes


def plot_dda():
    print("Enter coordinates for DDA Line:")
    x1 = int(input("x1: "))
    y1 = int(input("y1: "))
    x2 = int(input("x2: "))
    y2 = int(input("y2: "))

    xes, yes = dda_line(x1, y1, x2, y2)

    plt.figure(figsize=(6, 6))
    plt.plot(xes, yes, marker='*', linestyle='--', color='red')
    plt.title("DDA Line Drawing Algorithm")
    plt.xlabel("X-axis")
    plt.ylabel("Y-axis")
    plt.grid(True)
    plt.axis('equal')
    plt.show()
```
